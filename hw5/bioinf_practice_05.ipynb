### Annotation - sex and eye colour

Let's start by determining this person’s sex and eye colour. 

````python
def search_rs(file_name, rs_list):
    for rs in rs_list:
        !grep {rs} {file_name}
````

````python
eye_rs = ['rs12203592', 'rs12896399', 'rs12913832', 'rs16891982', 'rs6119471']
skin_rs = ['rs12913832', 'rs16891982', 'rs6119471', 'rs1426654', 'rs885479', 'rs1545397']

search_rs("raw_data/SNP_raw_v4_Full_20170514175358.txt", eye_rs)
````
